```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df=pd.read_excel('Data_train.xlsx')
train_df.head()
```

```python
test_df=pd.read_excel('Test_set.xlsx')
test_df.head()
```

```python
final_df = pd.concat([train_df, test_df], ignore_index=True)
final_df
```

```python
final_df['Date_of_Journey'] = pd.to_datetime(final_df['Date_of_Journey'])
final_df['Date']=final_df['Date_of_Journey'].dt.day
final_df['Month']=final_df['Date_of_Journey'].dt.month
final_df['Year']=final_df['Date_of_Journey'].dt.year
final_df.drop('Date_of_Journey',axis=1,inplace=True)
final_df.head()
```

```python
final_df.info()
```

```python
final_df['Arrival_Time']=final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df.head()
```

```python
final_df.isnull().sum()
```

```python
final_df['Arrival_Time']=pd.to_datetime(final_df['Arrival_Time'],format='%H:%M')
final_df['Arrival_Hour'] = final_df['Arrival_Time'].dt.hour
final_df['Arrival_Minute'] = final_df['Arrival_Time'].dt.minute
final_df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df['Dep_Time']=pd.to_datetime(final_df['Dep_Time'],format='%H:%M')
final_df['Dep_Hour'] = final_df['Dep_Time'].dt.hour
final_df['Dep_Minute'] = final_df['Dep_Time'].dt.minute
final_df.drop('Dep_Time', axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df['Source'].unique()
```

```python
final_df['Source']=final_df['Source'].map({'Banglore':1,'Kolkata':2,'Delhi':3,'Chennai':4,'Mumbai':5})
```

```python
final_df.head()
```

```python
final_df['Destination'].unique()
```

```python
final_df['Destination']=final_df['Destination'].map({'New Delhi':1,'Banglore':2,'Cochin':3,'Kolkata':4,'Delhi':5,'Hyderabad':6})
final_df.head()
```

```python
final_df['Route'].unique()
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'2 stops':1,'1 stop':2,'3 stops':3,'4 stops':4})
final_df.head()
```

```python
final_df.drop('Route',axis=1,inplace=True)
final_df.head()
```
